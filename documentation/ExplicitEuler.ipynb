# Explicit Euler 


The explicit Euler method can be described as unstable — that is, it is impossible to find a sufficiently small time step that ensures the error vanishes over infinite time.

Explicit Euler is formed is given by the following equation:


$$
y_{n+1} = y_n + h \cdot f(y_n)
$$

Where:  
- $y_n$ — current value of the variable  
- $y_{n+1}$ — next value of the variable  
- $h$ — time step  
- $f(y_n)$ — derivative of $y$ at step $n$


## Usage


In order to run the Explicit Euler case one should follow the $runmassspring.sh$ script and execute with following $euler$ method
