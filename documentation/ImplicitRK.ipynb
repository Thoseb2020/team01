# Implicit Range Kutta

The implicit Runge-Kutta method is another approach that performs well in computing stable solutions of ODEs. The algorithm is implemented as a three-step procedure.
- Predictor step:  
  $$k_1 = f(y_n)$$

- Midpoint evaluation:  
  $$k_2 = f\left(y_n + \frac{1}{2}\tau k_1\right)$$

- Update:  
  $$y_{n+1} = y_n + \tau k_2$$



# Usage

In order to run the Crank-Nicolson method one should execute the computational file and later the results parser.
Script let user to tune parameters
- period
- number of steps
- system i.e MS (mass-spring system) or RC system

Algorith computes results for three schemas:
- Gauss2 - Gauss-Legendre 2 stage
- Gauss3 - Gauss-Legendre 3 stage
- Radau

and outputs it's results to console

Generated data can be parsed by prepared python script
