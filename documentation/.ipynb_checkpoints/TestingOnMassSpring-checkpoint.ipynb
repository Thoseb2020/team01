# Testing on Mass-Spring system

Mass-spring system was choosen as one of the testing environment. We can estimate M-S system with no damping to work like mathematical oscillator - i.e energy, amplitude is conserved over time. 

Mass spring system is described be following relation:


$$
y_{0}' = y_1
$$

$$
y_{1}' = -\frac{k}{m} y_0
$$

and we should expect results that follows the following patten:


$$x(t) = Acos(\omega_0 t + \phi)$$
$$v(t) = -A\omega_0 sin(\omega_0 t + \phi)$$

where:

$$\omega = \frac{k}{m}$$
