# Implicit Euler

First solution that can generate a stable solution is Implicit Euler method, which is denoted by:

$$
y_{n+1} = y_n + h * f(t_{n+1}, y_{n+1})
$$

# Usage

In order to run the Explicit Euler case one should follow the $runmassspring.sh$ script and execute with following $improved$ method
