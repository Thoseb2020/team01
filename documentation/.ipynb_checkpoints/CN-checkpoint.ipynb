# Crank Nicolson

This approach uses 3 points stencil and can be expressed in a form:


$$
u^{n+1} = u^{n} + \frac{\Delta t}{2} \left( f(u^{n}) + f(u^{n+1}) \right)
$$


# Usage

In order to run the Crank-Nicolson method one should follow the $runmassspring.sh$ script and execute with following $CN$ method
