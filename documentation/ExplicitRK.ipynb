# Range Kutta

The explicit Runge-Kutta method is another approach that performs well in computing stable solutions of ODEs. The algorithm is implemented as a three-step procedure.
- Predictor step:  
  $$k_1 = f(y_n)$$

- Midpoint evaluation:  
  $$k_2 = f\left(y_n + \frac{1}{2}\tau k_1\right)$$

- Update:  
  $$y_{n+1} = y_n + \tau k_2$$
